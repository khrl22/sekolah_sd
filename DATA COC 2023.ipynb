```python
import pandas as pd

df = pd.read_csv('coc_sample.csv')
df.head()
df.info()
df.describe()
```

```python
df = df[['clan_tag', 'clan_name', 'clan_description', 'clan_location',
         'clan_level', 'clan_points', 'war_wins', 'war_losses',
         'war_ties', 'num_members', 'clan_war_league']]
```

```python
df.isnull().sum()
```

```python
df['clan_description'] = df['clan_description'].fillna('No description')
df['clan_location'] = df['clan_location'].fillna('Unknown')
df['clan_war_league'] = df['clan_war_league'].fillna('Unranked')
```

```python
df['clan_name'] = df['clan_name'].str.title()
df['clan_location'] = df['clan_location'].str.title()
df['clan_level'] = df['clan_level'].astype(int)
df['num_members'] = df['num_members'].astype(int)
```

```python
df.to_csv('coc_cleaned.csv', index=False)
print(" Data cleaned & saved.")
```
